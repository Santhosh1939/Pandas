```python
df=pd.DataFrame({'Name':['sam','ram','sravs','ram','sam'],"grade":['A','C','A','B','A'],"marks":['98','45','90','87','92']})
df
```

```python
df.T
```

```python
df
```

```python
df1 = pd.get_dummies(df['Name'], dtype=int)
df2 = pd.get_dummies(df['grade'], dtype=int)
print(df1)
print(df2)
```

```python
df
```

```python
frames = [df,df1,df2]
result = pd.concat(frames,axis=1)
print(result)
result.drop(['Name','marks'],axis=1,inplace=True)
# inplace=true will perform any action permanently
result
```
